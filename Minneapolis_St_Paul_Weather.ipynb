```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# Loading The Dataset
df = pd.read_csv('Weather.csv')

# Cleaning DataSet
df_clean = df.iloc[:, 0:2].copy()
df_clean = df[:-54750]
df_clean['Date'] = pd.to_datetime(df_clean['Date'])
df_clean['Date'] = df_clean['Date'].dt.strftime('%d')

```

```python
# Creating Bar Plot
plt.bar(df_clean['Date'], df_clean['Maximum Temperature degrees (F)'])
plt.title('Total Affected Individuals Over Time')
plt.xlabel('Date')
plt.ylabel('Maxinum Temperature Degrees(f)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Creating Stem Plot
plt.stem(df_clean['Date'], df_clean['Maximum Temperature degrees (F)'])
plt.title('The Maximum Temprature in the Twin Cites Area')
plt.xlabel('Date')
plt.ylabel('Maximum Temperature degrees (F)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
